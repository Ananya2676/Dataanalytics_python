 Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Mydata = {"Name":['Ram','Sam','Joe','Asha'],
          "Age":[23,22,26,47],
          "Salary":[1000,30000,30000,35000],
          "Experience":[2,1,3,10]
         }
```

```python
df = pd.DataFrame(Mydata)
df.head()
df.info()
```

1.Histogram

```python
plt.figure(figsize = (4,2))
sns.histplot(df['Salary'], kde = True, bins = 2)
plt.title("Distribution of Salary")
plt.show()
```

1.Negative Skew,lower salary value,below average value detected
2.no auterliers 
3.average salary is around 30000
4.majority salary values are between 200 and 35000 

Corelation Matrix(heatmap) 

```python
# Step 1:filter numerical data
ndf = df.select_dtypes(include = ['number'])
ndf.head()
```

```python
# Step 2:
plt.figure(figsize = (4,3))
sns.heatmap(ndf.corr(),cmap = "coolwarm",annot = True, linewidths= 0.5)
plt.title("Correlation between Age,Sal and Exp")
plt.show()
```

1.Age and Experience are more correlated to each other
2.Age and Salary is less correlated to each other

```python
plt.figure(figsize = (4,3))
sns.boxplot(x = df['Age'])
plt.title("Age Distribution")
plt.show()
```

1.one abnormal value is around 47
2.the average age is around 24

Countplot

```python
mydata1 = {"Name":['Ram','Sam','Joe','Asha'],
          "Age":[23,22,26,47],
          "Salary":[1000,30000,30000,35000],
          "Experience":[2,1,3,10],
           "G":['M','F','M','M']
         }
df1 = pd.DataFrame(mydata1)
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x = df1['Experience'], palette = 'bright', hue = df1['G'])
plt.title("count of Exp")
plt.show()
```

```python
mydata1 = {"Name":['Ram','Sam','Joe','Asha'],
          "Age":[23,22,26,47],
          "Salary":[1000,30000,30000,35000],
          "Experience":[2,3,3,10],
           "G":['M','F','M','M']
         }
df1 = pd.DataFrame(mydata1)
```
